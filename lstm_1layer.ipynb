```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time

from sklearn import linear_model

from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout, GRU, Bidirectional
from keras.optimizers import SGD
from keras import regularizers
from keras.regularizers import l1, l2

import math
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
def sgn(x):
    if x > 0:
        return (1)
    else:
        return (0)
```

```python
def sgn_f(x):
    if x > 0:
        return (0)
    else:
        return (1)
```

```python
path = "/Users/wuyue/Desktop/analysis/stock/"
files = os.listdir(path)
valid_file = []

for file in files:
    if file[-4:] == ".csv":
        valid_file.append(file)
        
print (len(valid_file))
```

```python
model = Sequential()

model.add(LSTM(100, input_shape = (1, 4), dropout=0.5))
model.add(Dense(1, activation = "sigmoid"))
model.compile(loss = "mean_squared_error", optimizer = "adam")

model.summary()
```

```python
start = time.clock()

mae = []
mse = []
counter = 1

true_5 = []
true_7 = []
true_14 = []
true_28 = []
true_60 = []

prediction_5 = []
prediction_7 = []
prediction_14 = []
prediction_28 = []
prediction_60 = []

for file in valid_file:
    start_s = time.clock()

    
    print ("training on", file)
    print (counter)
    
    file_name = path + file
    df = pd.read_csv(file_name).drop(columns=["Date", "OpenInt"])
    df = df[-500:]
    train = df[:-100]
    test = df[-100:]
    
    true_price = test["Close"].tolist()
    
    true_5_days = sgn(true_price[4]-true_price[0])
    true_5.append(true_5_days)
    true_7_days = sgn(true_price[6]-true_price[0])
    true_7.append(true_7_days)
    true_14_days = sgn(true_price[13]-true_price[0])
    true_14.append(true_14_days)
    true_28_days = sgn(true_price[27]-true_price[0])
    true_28.append(true_28_days)
    true_60_days = sgn(true_price[59]-true_price[0])
    true_60.append(true_60_days)
    
    c_mm = MinMaxScaler().fit(df.iloc[:, 3:4].astype('float64'))
    minmax = MinMaxScaler().fit(df.iloc[:, :].astype('float64'))

    df_log = minmax.transform(df.iloc[:, :].astype('float64'))
    df_log = pd.DataFrame(df_log)

    df_train = df_log.iloc[:-100]
    df_test = df_log.iloc[-100:]
    
    train_data = np.array(df_train.iloc[:, :-1])
    train_label = np.array(df_train.iloc[:, 4])
    test_data = np.array(df_test.iloc[:, :-1])
    test_label = np.array(df_test.iloc[:, 4])

    train_data_reshape = train_data.reshape((train_data.shape[0], 1, train_data.shape[1]))
    test_data_reshape  = test_data.reshape((test_data.shape[0], 1, test_data.shape[1]))
    
    training_data, validation_data, training_res, validation_res = train_test_split\
    (train_data_reshape, train_label, test_size = 0.15)

    trained_model = model.fit(training_data, training_res, epochs=30, batch_size=8,\
                              validation_data=(validation_data, validation_res), shuffle=False)
    
    
    
    plt.figure()
    plt.plot(trained_model.history['loss'], color = 'g', label = 'Training Loss')
    plt.plot(trained_model.history['val_loss'], color = 'r', label = 'Validation Loss')
    plt.title('Train/Validation Loss Function')
    plt.axis([0, 30, 0, 1.05*max(max(trained_model.history['val_loss']),max(trained_model.history['loss']))])
    plt.legend() 
    plt.show()

    
    prediction_mm = model.predict(test_data_reshape)
    res_mm = MinMaxScaler().fit(prediction_mm.astype('float64'))
    pre_mm = res_mm.fit_transform(prediction_mm)
    prediction = c_mm.inverse_transform(pre_mm)
    
    prediction_5_days = sgn(prediction[4]-prediction[0])
    prediction_5.append(prediction_5_days)
    prediction_7_days = sgn(prediction[6]-prediction[0])
    prediction_7.append(prediction_7_days)
    prediction_14_days = sgn(prediction[13]-prediction[0])
    prediction_14.append(prediction_14_days)
    prediction_28_days = sgn(prediction[27]-prediction[0])
    prediction_28.append(prediction_28_days)
    prediction_60_days = sgn(prediction[59]-prediction[0])
    prediction_60.append(prediction_60_days)
    
    plt.style.use("seaborn-darkgrid")
    plt.figure(figsize = (15, 5))
    plt.plot(test["Close"].values)
    plt.plot(prediction)
    plt.ylabel("Price")
    plt.legend(["True Trend", "Prediction"])
    plt.show()
    
    print ("MAE of", file, "is", mean_absolute_error(test["Close"].values, prediction))
    mae.append(mean_absolute_error(test["Close"].values, prediction))
    
    print ("MSE of", file, "is", mean_squared_error(test["Close"].values, prediction))
    mse.append(mean_squared_error(test["Close"].values, prediction))
    
    end_s = time.clock()
    
    print (len(prediction_5))
    print ("training time on", file, end_s-start_s)
    print ("===========================================")
    counter += 1
    
end = time.clock()
print ((end-start)/60, "min")
```

```python
print (accuracy_score(true_5,prediction_5))
print (accuracy_score(true_7,prediction_7))
print (accuracy_score(true_14,prediction_14))
print (accuracy_score(true_28,prediction_28))
print (accuracy_score(true_60,prediction_60))
```
